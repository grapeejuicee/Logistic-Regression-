# **LABEL ENCODING**

```python
#GET DATATYPES OF EACH COLUMN IN OUR DATASET
DF.dtypes
```

```python
#ENCODE THE DIGNOSIS VALUES


from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
DF['diagnosis'] = labelencoder.fit_transform(DF['diagnosis'])

```

```python
#DISPLAY DF
DF
```

# **SPLIT DATASET AND FEATURE SCALING**

```python
#SPLIT DATASET INTO INDEPENDENT AND DEPENDENT DATASET
X=DF.iloc[:,2:].values
Y=DF.iloc[:,1].values
```

```python
#SPLITING DATASETS INTO TRAINING(75%) AND TESTING(25%)
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
#SCALING THE DATA (FEATURE SCALING)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
#PRINT DATA

X_train

```

# **Build a logistic regression model**

```python
#build a logistic regression classifier
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,Y_train)
```

```python
#make use of test data to make predition on the test data
prediction = classifier.predict(X_test)
```

# **Performance evaluation**

```python
#plot confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm=confusion_matrix(Y_test,prediction)
print(cm)
sns.heatmap(cm,annot=True)
```

```python
#get accuracy score for the model
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,prediction))
```
